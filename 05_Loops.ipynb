```python
# Prime Number COde

n = int(input("Enter a number : "))

for i in range(2,n//2 + 1):
    if n%i == 0:
        print(n,'is not Prime')
        break
else:
    print(n,'is Prime')
```

```python
# Prime numbers within a range

low = 2
high = 20

for n in range(low, high + 1):
    for i in range(2,n//2 +1):
        if n%i == 0:
            break
    else:
        print(n)
```

### List comprehension with condition filtering

```python
# Odd numbers
a = []
for i in range(5):
    if i%2 != 0:
        a.append(i)
print(a)
```

```python
# List comprehension
a = [i for i in range(5) if i%2 != 0]
print(a)
```

```python
# Even numbers
a=[]
for i in range(9):
    if i%2==0:
        a.append(i)
print(a)
```

```python
# List comprehension
l=[i for i in range(9) if i%2==0]
l
```

```python
# Both Even and ODD numbers
a = []
for i in range(5):
    if i%2 != 0:
        a.append((i,'ODD'))
    else:
        a.append((i,'EVEN'))
print(a)
```

```python
# LC

a = [(i,'ODD') if i%2 != 0 else (i,'EVEN') for i in range(5)]
print(a)
```

## Question

Given a string and a width you have to print the string line by line with maximum characters in a line equals to the width mentioned

```python

a = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
w = 4
# output:               # ABCD
                        # EFGH
                        # IJKL
                        # MNOP
                        # QRST
                        # UVWX
                        # YZ

b = []
for i in range(0,len(a),w):
    b.append(a[i:i+w])
print(b)

b = '\n'.join(b)
print(b)
```

```python
#LC

a = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
w = 4

b = '\n'.join([a[i:i+w] for i in range(0,len(a),w)])
print(b)
```

### Remove duplicate items

```python
a = [10,40,40,40,20,30,40,50,50]
b=[]
for i in a:
    if a.count(i)<=1:
        b.append(i)
print(b)

c=[i for i in a if a.count(i)<=1]
c
```

```python
# Remove duplicates from  a list

a = [10,40,40,40,20,30,40]
b = []
for i in a:
    if a.count(i) == 1:
        b.append(i)
print(b)

a = [i for i in a if a.count(i) == 1]
print(a)
```
