<h2 style="color:#BA4A00" align ="center">.seek() method</h2>
<h3 style="color:navy">In Python, seek() function is used to change the position of the File Handle to a given specific position. File handle is like a cursor, which defines from where the data has to be read or written in the file.</h3>

```python
f = open('test.txt')
a = f.read()
f.seek(5)# Now cursor will read from 10th index ..
b = f.read()
f.close()
```

```python
a
```

```python
b
```

<h2 style="color:#BA4A00" align ="center">.tell() method</h2>
<h3 style="color:navy">In Python, . tell() method can be used to get the position of File Handle(or cursor). tell() method returns current position of file object. This method takes no parameters and returns an integer value..</h3>

```python
f = open('test.txt')
a = f.read()
print("Length of our file :",len(a))
print(f.tell())# Position of Cursor after reading all elements of file 

f.seek(5)# Now Seek will bring corsur to 0th index 
print(f.tell())
b = f.read()
f.close()
```

```python
print(a)
```

```python
print(b)
```

<h2 style="color:#BA4A00" align ="center">New way of handling file .  . </h2>

```python
with open('test.txt','r') as f: # here f is just a name 
    print(f.read())
```

<h3 style="color:green"> in above code ,open () function takes one argument and that is path of file or file name</h3>
<h3 style="color:green">and  'with' keyword close the file once access to it is no longer needed.</h3>


 That is the reason why we call only open() not close()

```python
with open('test.txt','r') as f:
    for line in f:
        print(line)
```

```python
with open('test.txt','r') as f:
    lines=f.readlines()
    print(lines)
```

```python
lines[::2]
```

```python
for line in lines:
    print(line)
```

```python
for line in lines:
    print(line,end=" ")
```

```python
for line in lines:
    print(line.strip())
```

```python
#Writing multiple lines . 
path="test.txt"
with open(path,'w') as f:
    f.write("Phla phla line hai  .\n")
    f.write(("aur ye dusra line hai."))
```

```python
path="test.txt"
with open(path) as f:
    lines = f.readlines()
for line in lines:
    print(line)
```

```python
path="test.txt"
with open(path,'w') as f:
    f.write("\nPhla phla line hai  .\n")
    f.write(("aur ye dusra line hai.\n"))
    f.write("third line hai  .\n")
    f.write(("fourth line hai.\n"))
```

```python
path="test.txt"
with open(path) as f:
    lines = f.readlines()
for line in lines:
    print(line)
```

```python
#Appending multiple line
path="test.txt"
with open(path,'a') as f:
    f.write("Phla phla line hai  .\n")
    f.write(("aur ye dusra line hai.\n"))
    f.write("third line hai  .\n")
    f.write(("fourth line hai.\n"))
```

```python
path="test.txt"
with open(path) as f:
    lines = f.readlines()
for line in lines:
    print(line,end="")
```
