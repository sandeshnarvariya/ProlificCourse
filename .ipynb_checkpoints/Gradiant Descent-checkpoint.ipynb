# Gradient Descent   :

#### Gradient descent is one of the simplest and widely used algorithms in machine learning, mainly because it can be applied to any function to optimize it.

#### Gradient descent is an iterative optimization meta algorithm to find the minimum of a function. Here that function is our Loss Function 0r Cost Function 0r MSE.

#### Actual G0al of  Gradient Descent  is to find global mimina of Cost Function. So that our model can get 'Best Fit Line' .

## How  Gradient Descent works : 

#### Imagine a valley and a person with no sense of direction who wants to get to the bottom of the valley. He goes down the slope and takes large steps when the slope is steep and small steps when the slope is less steep. He decides his next position based on his current position and stops when he gets to the bottom of the valley which was his goal.

#  steps to perform Grad_Desc -: 

# Step 1: Get Derivatives of m & b


# step 2: (update m & b)


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Example 1

```python
def grad_desc(x,y):
    m_curr=0
    b_curr=0
    lr=0.01
    itr=10000
    n=len(x)
    for i in range(itr):
        y_p=m_curr*x+b_curr
        mse=1/n*(sum([val**2 for val in (y-y_p)]))
        md=-2/n*(sum(x*(y-y_p)))
        bd=-2/n*(sum(y-y_p))
        m_curr=m_curr-lr*md
        b_curr=b_curr-lr*bd
        print(f"for {i} iteration =\n m_curr :{m_curr} , b_curr :{b_curr} ,\nmse :{mse}")
        print()
x = np.arange(2,11,2)
y = np.arange(5,18,3)
grad_desc(x,y)#calling our function
```

# example 2 : (With datasets)

```python
df =pd.read_csv(r'Datasets\areaprice.csv')
df.sample(3)
```

```python
def grad_desc(x,y):
    m_curr=0
    b_curr=0
    lr=0.001
    itr=200
    n=len(x)
    for i in range(itr):
        y_p=m_curr*x+b_curr
        mse=1/n*(sum([val**2 for val in (y-y_p)]))
        md=-2/n*(sum(x*(y-y_p)))
        bd=-2/n*(sum(y-y_p))
        m_curr=m_curr-lr*md
        b_curr=b_curr-lr*bd
        print(f"for {i} iteration =\n m_curr :{m_curr} , b_curr :{b_curr} ,\nmse :{mse}")
        print()

grad_desc(df.area,df.price)#calling our function
```
