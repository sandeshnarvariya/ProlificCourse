### try-except-else block

```python

try:
    n = int(input("Enter a number : "))
    a = 1/n
    print(a)
except:
    print("ERROR HANDLED")
else:
    print(f'square of your number {n} is {n**2}')
    
print("FURTHER CODE")
```

### Try-finally block

finally block will execute irrespective of any error or not

```python
try:
    n = int(input("Enter a number : "))
    a = 1/n
    print(a)
finally:
    print("No matter what i will defenetly run .")
    
print("FURTHER CODE")
print("FURTHER CODE")
```

### try-except-finally block

```python
try:
    n = int(input("Enter a number : "))
    a = 1/n
    print(a)
except:
    print("ERROR HANDLED")
finally:
    print("FINALLY")
    
print("FURTHER CODE")
```

### try-except-else-finally block

```python
try:
    n = int(input("Enter a number : "))
    a = 1/n
except:
    print("ERROR HANDLED")
else:
    print('Reciprocal of your number {} is {}'.format(n,a))
finally:
    print("FINALLY")
print("FURTHER CODE")
```

<h2 style="color:navy" align ="center">sys.exc_info </h2>
<h3 style="color:navy" align ="center"> Return information about the most recent exception caught by try block</h3>

```python
import sys

help(sys.exc_info)
```

```python
import sys

a = [10,'a',0,[1,2,3]]

for i in a:
    print("Entry :",i)
    try:
        n = 1/int(i)
    except:
        print("Error :",sys.exc_info()[0])

    else:
        print("Reciprocal :",n)
    print()
```

### User defined Error

```python
try:
    name=input("Enter your name in capital letters only :")
    raise NameError("Not in Capital ")  # Raise Error 
except NameError: 
    print ("Dont use small letters")
    raise  # To determine whether the exception was raised or not 
```

```python
def positive(x):
    if x>=0:
        raise Exception ('Number already positive')
    else:
        return abs(x)
    
print(positive(-6))
print(positive(5))
```

```python
#Raising Exception

def lift(floor):
    if floor <0:
        raise Exception ("We don't provide service in PATAL LOOOK")
    return floor

for i in range(5):
    try:
        a = int(input("Enter your Floor : "))
        b = lift(a)
        print(f"WE ARE HEADING TO {b}th FLOOR:")
    except :#Exception as e:
        print("Error in your floor :")
        raise
```

### Assert

The assert keyword lets you test if a condition in your code returns True, if not, the program will raise an AssertionError.

```python
x = input()

#if condition returns True, then nothing happens:
assert x == "hello"

#if condition returns False, AssertionError is raised:
assert x == "goodbye"
```

### Write a message if the condition is False:

```python
x = "goodbye"

#if condition returns False, AssertionError is raised:
assert x == "hello", "x should be 'goodbye'"
```

```python
def positive(x):
    assert x<0, " Number already positive"
    return abs(x)

print(positive(5))
print(positive(-6))
```
